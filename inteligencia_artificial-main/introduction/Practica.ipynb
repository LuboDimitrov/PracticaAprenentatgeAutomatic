## Conversió dels strings a float

Hem de convertir a un valor numèric les columnes que tenen caràcters com €,K,M ja que el nostre model no enten el caràcters no numèrics.

```python
def value_to_float(x):
    """
    From K and M to float.
    
    """
    x = x.replace('€', '')
    ret_val = 0.0
    
    if type(x) == float or type(x) == int:
        ret_val = x
    if 'K' in x:
        if len(x) > 1:
            ret_val = float(x.replace('K', ''))
        ret_val = ret_val *1000
    if 'M' in x:
        if len(x) > 1:
            ret_val = float(x.replace('M', ''))
        ret_val = ret_val * 1000000.0
    return ret_val
```

Aplicam aquesta funció a cada valor de la columna Value i a continuació ho aplicarem a altres columnes com el salari o la clàusula

```python
df["Value"] = df["Value"].apply(value_to_float)
df.head()
```

```python
df["Wage"] = df["Wage"].apply(value_to_float)
df.head()
```

```python
df["Release Clause"] = df["Release Clause"].apply(value_to_float)
df.head()
```

## Canviar les columnes que són sumes a un sol valor (LS,etc)

Per fer-ho he fet una funció que suma els dos valors separats per el símbol "+" 

```python
def suma_valors(x):
    if x == x:
        if type(x) == float or type(x) == int:
            ret_val = x
        value = x.split('+')
        ret_val = int(value[0]) + int(value[1])
        return ret_val 
```

## Convertim l'estatura en cm

Necessitam convertir l'estatura ja que en el dataset original l'estatura està amb valors no numèrics i per poder fer una bona predicció feim la conversió.

```python
def height_to_cm(x):
    value = str(x)
    height = value.split("'")
    feet = int(height[0])
    inch = int(height[1])
    total = feet*12 + inch
    cm = total * 2.54
    return cm
```

Aplicam aquesta funció a la columna de l'estatura i ara ja estarà en un valor numèric que el nostre algorisme entendrà.

```python
df["Height"] = df["Height"].apply(height_to_cm)
df.head()
```

## One hot encoding
Per a que l'algoritme de ML entengui ses variables no numèriques.
Per exemple en el cas del club, es crearan tantes columnes adicionals com clubs hi hagi i si per exemple miram el cas den Messi, la columna del club FC Barcelona valdrà 1 i tota la resta valdran 0. Així el nostre algorisme entèn que el jugador Leo Messi pertany al club FC Barcelona

Les columnes a les quals aplicam el one hot encoding són:

- Club
- Body Type
- Position
- Preferred Foot
- Work rate

```python
clb = df.pop("Club")

df = pd.concat([df.reset_index(drop=True), pd.get_dummies(clb, prefix='clb').reset_index(drop=True)], axis=1, sort=False)
df.head()
```

```python
bt = df.pop("Body Type")

df = pd.concat([df.reset_index(drop=True), pd.get_dummies(bt, prefix='bt').reset_index(drop=True)], axis=1, sort=False)
df.head()
```

```python
pos = df.pop("Position")

df = pd.concat([df.reset_index(drop=True), pd.get_dummies(pos, prefix='pos').reset_index(drop=True)], axis=1, sort=False)
df.head()
```

```python
pf = df.pop("Preferred Foot")

df = pd.concat([df.reset_index(drop=True), pd.get_dummies(pf, prefix='pf').reset_index(drop=True)], axis=1, sort=False)
df.head()
```

```python
wr = df.pop("Work Rate")

df = pd.concat([df.reset_index(drop=True), pd.get_dummies(wr, prefix='wr').reset_index(drop=True)], axis=1, sort=False)
df.head()
```

## Normalització

Hem de normalitzar les dades perquè necessitam fer feina sobre el mateix rang per obtenir una bona predicció.
Per normalitzar un valor hem de restar la mitjana a aquest valor i després dividir entre la desviació típica.
Les columnes que hem de normalitzar són les columnes resultants de llevar les columnes innecessàries menys la del valor.
Tampoc necessitam les columnes resultants de fer el one hot encoding ja que tenen valors entre 0 i 1

```python
columnes =  ['Age','Overall','Potential','Wage','International Reputation','Weak Foot','Skill Moves','Height','Weight','LS','ST',
        'RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB',
        'CB','RCB','RB','Crossing','Finishing','HeadingAccuracy','ShortPassing','Volleys','Dribbling','Curve','FKAccuracy',
        'LongPassing','BallControl','Acceleration','SprintSpeed','Agility','Reactions','Balance','ShotPower','Jumping',
        'Stamina','Strength','LongShots','Aggression','Interceptions','Positioning','Vision','Penalties','Composure','Marking',
        'StandingTackle','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes','Release Clause']
```

```python
for i in columnes:
    df[i] = (df[i] - X_train[i].mean())/X_train[i].std(ddof=0)
df
```

## R2 Score

R2 score és una mesura que s'empra per predir futurs resultats o per demostrar una hipòtesi. Aquest coeficient pot adquirir valors entre qualsevol valor negatiu i 1 i determina la qualitat del model per replicar els resultats. Ens interessa que sigui el més proper a 1.

```python
r2_score(preds, y_test)
```

Obtenim el resultat de 0.966 el qual ens indica que el nostre model és bastant precís.

## Mean absolute error

Serveix per quantificar la precisió la precisió d'una tèncnica de predicció comparant els valors predits amb els valors observats. Al tractar-se d'una mètrica d'error ens interessa que sigui el més baix possible. 



```python
mean_absolute_error(preds, y_test)
```
